```python
def genre_based_recommendation(user_genre, num_recommendations=5):
    genre_movies = movies_data[movies_data['genres'].str.contains(user_genre, case=False, na=False)]
    genre_movies = genre_movies.sort_values(by='popularity', ascending=False)
    recommendations = genre_movies.head(num_recommendations)
    return recommendations[['title', 'genres', 'popularity']]
user_preferred_genre = 'Action'
genre_recommendations = genre_based_recommendation(user_preferred_genre)
print(f"Genre-based recommendations for '{user_preferred_genre}':")
print(genre_recommendations)
```

```python
def popularity_based_recommendation(num_recommendations=10):
    popular_movies = movies_data.sort_values(by='popularity', ascending=False)
    recommendations = popular_movies.head(num_recommendations)
    return recommendations[['title', 'popularity']]
popularity_recommendations = popularity_based_recommendation()
print("Popularity-based recommendations:")
print(popularity_recommendations)
```

```python
def budget_based_recommendation(max_budget, num_recommendations=5):
    budget_movies = movies_data[movies_data['budget'] <= max_budget]
    budget_movies = budget_movies.sort_values(by='budget', ascending=False)
    recommendations = budget_movies.head(num_recommendations)
    return recommendations[['title', 'budget']]
max_budget_limit = 4000000
budget_recommendations = budget_based_recommendation(max_budget_limit)
print(f"Budget-based recommendations for a maximum budget of {max_budget_limit}:")
print(budget_recommendations)
```
